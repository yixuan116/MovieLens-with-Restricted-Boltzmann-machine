# Baseline Evaluation

Run baseline recommenders under the same train/test split.

```python
import os
import sys

sys.path.append(os.path.abspath("src"))

from data_prep import DataConfig, load_movielens
from baselines import PopularityBaseline, MatrixFactorizationBaseline
from metrics import evaluate_model

config = DataConfig(data_dir="data/movielens-20m", max_users=2000, max_items=4000)
data = load_movielens(config)
train_matrix = data["train_matrix"]
test_matrix = data["test_matrix"]

pop = PopularityBaseline()
pop.fit(train_matrix)
print("Popularity:", evaluate_model(pop.predict_scores, test_matrix, train_matrix, k=20))

mf = MatrixFactorizationBaseline(n_components=64)
mf.fit(train_matrix)
print("Matrix Factorization:", evaluate_model(mf.predict_scores, test_matrix, train_matrix, k=20))
```
